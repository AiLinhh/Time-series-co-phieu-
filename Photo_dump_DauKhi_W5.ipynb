```python
#1. PhÃ¢n tÃ­ch level:

path = 'Downloads/XuLyDaukhixyz.xlsx'
df = pd.read_excel(path)
df
#TÃ­nh trung bÃ¬nh Ä‘á»™ng (moving average) cá»§a biáº¿n PCG trong 7 ngÃ y
df['PCG_MA7'] = df['PCG'].rolling(window=7).mean()
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df['Date'], df['PCG'])
ax.plot(df['Date'], df['PCG_MA7'])
ax.legend(['PCG', 'PCG_MA7'])
ax.set_title('Moving Average PCG')
plt.show()

#2. PhÃ¢n tÃ­ch trend:

from sklearn.linear_model import LinearRegression

X = df.index.values.reshape(-1, 1)
y = df['PCG']
model = LinearRegression().fit(X, y)
trend = model.predict(X)
df['PCG_trend'] = trend

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df['Date'], df['PCG'])
ax.plot(df['Date'], df['PCG_trend'])
ax.legend(['PCG', 'PCG_trend'])
ax.set_title('Trend')
plt.show()

from statsmodels.tsa.seasonal import seasonal_decompose
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

result = seasonal_decompose(df['PCG'], model='multiplicative', period=30)
df['PCG_trend'] = result.trend
df['PCG_seasonality'] = result.seasonal
df['PCG_residual'] = result.resid
fig, ax = plt.subplots(nrows=4, ncols=1, figsize=(10, 12), sharex=True)
ax[0].plot(df['PCG'])
ax[0].set_ylabel('PCG')
ax[1].plot(df['PCG_trend'])
ax[1].set_ylabel('Trend')
ax[2].plot(df['PCG_seasonality'])
ax[2].set_ylabel('Seasonality')
ax[3].plot(df['PCG_residual'])
ax[3].set_ylabel('Residual')
plt.show()
```

```python
#PCG	PLX	PVB	PVC	PVO
path = 'Downloads/XuLyDaukhixyzz.xlsx'
df = pd.read_excel(path)


df['Date'] = pd.to_datetime(df['Date'])
# calculate the monthly percentage change
df['Pct_Change'] = df.groupby(['Year','Month'])['PCG'].pct_change()

# fill missing values with the mean of that month
df['Pct_Change'] = df.groupby(['Month'])['Pct_Change'].fillna(df['Pct_Change'].mean())

# create the pivot table with columns as months and rows as years
pivot = df.pivot_table(index='Year', columns='Month', values='Pct_Change', aggfunc='sum')

# plot the heatmap with green for positive changes and red for negative changes
import seaborn as sns
import matplotlib.pyplot as plt

cmap = sns.diverging_palette(10, 220, as_cmap=True)



fig = plt.figure(figsize=(17, 7))
sns.heatmap(pivot, cmap=cmap, center=0, annot=True, fmt=".2%")
plt.title("Percentage Change in PCG by Month and Year")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

path = 'Downloads/XuLyDaukhixyzz.xlsx'
df = pd.read_excel(path)
# Táº¡o pivot table
df_pivot = pd.pivot_table(df, values='PCG', index=['Month'], columns=['Year'])

# TÃ­nh moving average
df_pivot_ma = df_pivot.rolling(window=3).mean()

# Váº½ biá»ƒu Ä‘á»“ line chart káº¿t há»£p cáº£ dá»¯ liá»‡u gá»‘c vÃ  dá»¯ liá»‡u Ä‘Ã£ tÃ­nh moving average
sns.set_style("whitegrid")
sns.set_palette("husl")
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_pivot_ma, color="lightgray")
plt.title("PCG by Year")
plt.xlabel("Month")
plt.ylabel("PCG")
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
